```
# Handling missing values:

import pandas as pd

# sample dataset with missing values 
data = {
    "Customer" : ["Rahul", "Sneha", "John", "Kiran"],
    "City"     : ["Mumbai", "Delhi", None, "Pune"],
    "Age"      : [29, None, 45, 28],
    "Monthly_Bill": ["300", "500", "1000", None]
}

df = pd.DataFrame(data)

print("\n Original Data:")
print(df)


# handling missing values,
#(1) City  --> Unknown 
#(2) Age -> Mean of the  Age 
#(3) Monthly_Bill --> "0" zero 

df.fillna({     # Fillna() helps to put missing values data to specific data

    "City": "Unknown",
    "Age" : df["Age"].mean(),  #mean() mean of the Age 
    "Monthly_Bill": "0"
}, inplace = True)

# Convert string to integer 
df["Monthly_Bill"] = df["Monthly_Bill"].astype(int)  #astype() method helps to convet into int/float etc

#Print Cleaned data
print("\n ******  === Cleaned data ====== **** ")
print(df)

# Use Joins on the Dataframe - (i.e Merge )
plan_data = {
    "Monthly_Bill" : [300, 500, 100, 0],
    "Plan_Name" : ["Basic", "Premium", "Enterprise", "No Plan"]
}

df_plan = pd.DataFrame(plan_data)

print(plan_data)

# Merge the dataframe 
 
merged_df = pd.merge(df, df_plan, on="Monthly_Bill", how="left")  #pd.merge()  helps to to merge / join 

print("\n ***** ======  Print merged_df dataset  ====== ****")
print(merged_df)

```

```
# Feature engineering 


from snowflake.snowpark import Session #snowpark - inbuilt library to work with data in snowflake 
from snowflake.snowpark.functions import col, when

import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

import pandas as pd

# Create Snowflake session  - main entry point 
session = Session.builder.getOrCreate()

###====================================================
### Creating Sample dataset 
###=============================================================

def create_sample_date():    # def -> used to create function()
    data = {
        "PRICE": [100, 200, 300, 400],
        "QUANTITY" : [2, 5, 10, 15]
    }
    return pd.DataFrame(data)

#Create sample Dataset 
pdf = create_sample_date() 

# convert to Snowpark DataFrame 
df = session.create_dataframe(pdf)

print("Original data:")
df.show()

#-----------------------------------------
#Feature Engineering 
#------------------------------------------------------


df_fe = (
    df
    .with_column("REVENUE", col("PRICE") * col("QUANTITY"))
    .with_column("BIG_ORDER", when(col("QUANTITY")> 10, 1).otherwise(0))
)

print("\nAfter Feature engineering:")
df_fe.show()



```

```
# Predictive Model - Linear Regression 

import matplotlib.pyplot as plt
from scipy import stats

x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
